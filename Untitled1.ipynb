```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# List of image file paths
image_paths = ['one.jpg', 'two.jpg', 'three.jpg','four.jpg','five.jpg','six.jpg','seven.jpg','eight.jpg'] 
```

```python
for image_path in image_paths:
    # Load the thermal image
    thermal_image = cv2.imread(image_path, cv2.IMREAD_COLOR)
    
```

```python
# Display the image
cv2.imshow('Thermal Image', thermal_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Continue with the rest of your image processing code
thermal_gray = cv2.cvtColor(thermal_image, cv2.COLOR_BGR2GRAY)
thermal_gray_equalized = cv2.equalizeHist(thermal_gray)
    
```

```python
# Apply adaptive thresholding
thresholded_image = cv2.adaptiveThreshold(thermal_gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
   
```

```python
 # Find contours in the thresholded image
contours, _ = cv2.findContours(thresholded_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    
```

```python
# Draw contours on a copy of the original image
thermal_contours = thermal_image.copy()
cv2.drawContours(thermal_contours, contours, -1, (0, 255, 0), 2)

thermal_smoothed = cv2.GaussianBlur(thermal_gray, (5, 5), 0)
thermal_smoothed = cv2.medianBlur(thermal_gray, 5)

    
```

```python
# Display and save processed images
plt.subplot(131), plt.imshow(thermal_image[:, :, ::-1]), plt.title('Original Image')
plt.subplot(132), plt.imshow(thresholded_image, cmap='gray'), plt.title('Thresholded Image')
plt.subplot(133), plt.imshow(thermal_contours), plt.title('Contours')
plt.show()
```

```python
# Save processed images
processed_image_path = image_path.replace('.jpg', '_processed.jpg')
cv2.imwrite(processed_image_path, thermal_contours)
```

```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model

# Load the trained model
model = load_model('your_model.h5')  # Update with the path to your trained model


```

```python
# Preprocess the input image (similar to your training data preprocessing)
input_image = cv2.imread('two.jpg')  # Update with the path to your input image

```

```python
input_image = cv2.resize(input_image, (128, 128))  # Resize the image
input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)  # Convert to RGB format
input_image = input_image / 255.0  # Normalize pixel values
```

```python

input_image = input_image.reshape(1, 128, 128, 3)  # Adjust dimensions to (1, 128, 128, 3)

# Make predictions
predictions = model.predict(input_image)

# Define a threshold (e.g., 0.5)
threshold = 0.5

# Check if the probability for the "cracked" class is greater than the threshold
if predictions[0][0] > threshold:
    result = "Cracked"
else:
    result = "Not Cracked"

# Display the image
plt.imshow(input_image[0])
plt.title(f"Prediction: {result}")
plt.show()
```
