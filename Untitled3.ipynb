```python
import pandas as pd
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
```

```python
# Load the Excel dataset
df = pd.read_csv('crack2.csv')
```

```python
# Create empty lists to store images and labels
images = []
labels = []

# Iterate through the dataset to load and preprocess the images
for index, row in df.iterrows():
    image_path = row['Image File Path']
    label = row['Labels']

    # Load and preprocess the image
    image = cv2.imread(image_path)
    image = cv2.resize(image, (128, 128))  # Resize the images to a common size
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB format
    image = image / 255.0  # Normalize pixel values to [0, 1]

    images.append(image)
    labels.append(label)

# Convert the lists to NumPy arrays
images = np.array(images)
labels = np.array(labels)
```

```python
X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a label encoder
label_encoder = LabelEncoder()

# Encode the labels
df['Labels'] = label_encoder.fit_transform(df['Labels'])
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.preprocessing import LabelEncoder

# Use LabelEncoder to encode the labels
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)
```

```python
# Define and compile the model
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    layers.MaxPooling2D(2, 2),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(2, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train_encoded, epochs=10, validation_data=(X_test, y_test_encoded))
```

```python
label_encoder.classes_
```

```python
from sklearn.metrics import classification_report, precision_score, recall_score, accuracy_score

# Make predictions on the test data
y_pred = model.predict(X_test)

# Convert predicted probabilities to class labels (0 or 1)
y_pred_labels = np.argmax(y_pred, axis=1)
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test_encoded, y_pred_labels)

# Calculate precision and recall for each class
precision = precision_score(y_test_encoded, y_pred_labels, average=None)
recall = recall_score(y_test_encoded, y_pred_labels, average=None)

# Generate a classification report
class_report = classification_report(y_test_encoded, y_pred_labels, target_names=label_encoder.classes_)

```

```python

print(f'Accuracy: {accuracy}')
print(f'Precision for class "cracked": {precision[0]}')
print(f'Precision for class "not cracked": {precision[1]}')
print(f'Recall for class "cracked": {recall[0]}')
print(f'Recall for class "not cracked": {recall[1]}')
print('Classification Report:\n', class_report)
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the thermal image
image_path = 'DJI_20231216103516_0005_T.JPG'
thermal_image = cv2.imread(image_path)
```

```python
# Convert the image to grayscale
thermal_gray = cv2.cvtColor(thermal_image, cv2.COLOR_BGR2GRAY)

# Apply adaptive thresholding
thresholded_image = cv2.adaptiveThreshold(thermal_gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
```

```python
# Find contours in the thresholded image
contours, _ = cv2.findContours(thresholded_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw contours on a copy of the original image
thermal_contours = thermal_image.copy()
cv2.drawContours(thermal_contours, contours, -1, (0, 255, 0), 2)
```

```python
# Convert the image to grayscale
thermal_gray = cv2.cvtColor(thermal_image, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur to reduce noise
thermal_blurred = cv2.GaussianBlur(thermal_gray, (5, 5), 0)

# Define lower and upper bounds for the bright orange-red color (adjust these values as needed)
lower_bound = np.array([0, 100, 150], dtype=np.uint8)
upper_bound = np.array([40, 255, 255], dtype=np.uint8)

# Convert the image to HSV color space for color filtering
thermal_hsv = cv2.cvtColor(thermal_image, cv2.COLOR_BGR2HSV)

# Threshold the image to get a mask for the bright orange-red color
mask = cv2.inRange(thermal_hsv, lower_bound, upper_bound)

# Perform morphological operations to enhance the structures
kernel = np.ones((5, 5), np.uint8)
mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)

# Find contours in the mask
contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw contours on a black background
contour_image = np.zeros_like(thermal_image)
cv2.drawContours(contour_image, contours, -1, (255, 255, 255), thickness=cv2.FILLED)

# Apply the mask to the original image
result_image = cv2.bitwise_and(thermal_image, contour_image)

# Display the result
plt.imshow(result_image[:, :, ::-1])
plt.title("Detected Structures")
plt.show()
```

```python
# Drawing bounding boxes around significant contours
result_image = thermal_image.copy()  # Use the original image 'img' instead of 'thermal_image'
for contour in contours:  # 'contours' is the list of significant contours found in the corrosion detection code
    x, y, w, h = cv2.boundingRect(contour)
    cv2.rectangle(result_image, (x, y), (x + w, y + h), (0, 0, 0), 2)  # Set color to black (0, 0, 0)

# Display the result
cv2.imshow("Crack Detection", cv2.resize(result_image, (1000, 800)))
cv2.waitKey(0)
cv2.destroyAllWindows()
```
